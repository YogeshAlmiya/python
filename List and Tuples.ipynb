```python
#Code to check execution time for list
import timeit
code_to_test = """

listObj = ['Riti', 31, 'London', 78.88]
 
for elem in listObj:
    print(elem)

"""
elapsed_time = timeit.timeit(code_to_test, number=100)/100
print(elapsed_time)
```

```python
#Code to check execution time for tuples

code_to_test = """

tupleObj = ('Riti', 31, 'London', 78.88)
 
for elem in tupleObj:
    print(elem)
"""
elapsed_time = timeit.timeit(code_to_test, number=100)/100
print(elapsed_time)
```
